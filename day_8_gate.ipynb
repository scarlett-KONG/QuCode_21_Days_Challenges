The NOT gate in quantum computing is the **Pauli-X gate**, and its matrix is
$$X = 
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$
It maps $|0\rangle \to |1\rangle$
It maps $|1\rangle \to |0\rangle$
The Pauli-X (NOT) gate acting on a general qubit state is :
$$X \bigl( \alpha |0\rangle + \beta |1\rangle \bigr)  = \beta |0\rangle + \alpha |1\rangle $$


```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_hinton
import matplotlib.pyplot as plt
import numpy as np

#Define your custom state: alpha |0> + beta |1>
alpha = 1/np.sqrt(5)
#beta = 2/np.sqrt(5)
beta = np.sqrt(4/5)*np.exp(1j * np.pi / 4)   # Example with a phase
state = [alpha, beta]

# Create the statevector
state = Statevector([alpha, beta])
```

```python
from qiskit.quantum_info import Operator
from qiskit.quantum_info import Pauli
# Define the Pauli-X operator
pauli_x = Operator(Pauli('X'))
# Apply the Pauli-X gate
display(pauli_x.draw("latex"))
display(state.draw("latex"))
new_state_x = state.evolve(pauli_x)
display(new_state_x.draw("latex"))
display(plot_bloch_multivector(state))
display(plot_bloch_multivector(new_state_x))
display(plot_state_hinton(state))
display(plot_state_hinton(new_state_x))
```

# Multiple qubits system and Bell state

how we create a Bell state (entangled state) using two qubits

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import HGate, CXGate
qc = QuantumCircuit(2)
qc.append(HGate(), [0])  # Apply Hadamard to qubit 0
qc.append(CXGate(), [0, 1])  # Apply CNOT with control qubit 0 and target qubit 1
qc.draw("mpl")
```

Now we will add the measurement and then do the simulation and analysis the result.

```python
qc = QuantumCircuit(2,2)
qc.h(0)  
qc.cx(0, 1) 
qc.measure([0,1], [0,1])
qc.draw("mpl")

```
